```python
import pandas as pd

df = pd.read_csv('../data/sales.csv')
df
# select
df[['product', 'price']]
# where
df[df['category'] == 'Electronics']

# group by
df['revenue'] = df['price'] * df['quantity']

df.groupby('category')['revenue'].sum().reset_index()
# order by
df.groupby('category')['revenue'].sum().sort_values(ascending=False)






```
